# Agrupando

```python
groupping = pd.DataFrame()
groupping = pd.concat([mitochondrial, circulatory, cardiac_functioning, development])
groupping
```

```python
groupping['ID'] = groupping['ID'].str.replace('_', ' ')
groupping['ID'] = groupping['ID'].str.replace('Go', '')
groupping['ID'] = groupping['ID'].str.title()

groupping
```

```python
import seaborn as sns
import matplotlib.pylab as plt

ax = sns.barplot(data=groupping, x='ID', y='log')
plt.xticks(rotation=90)


def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .35)
ax.set_xlabel("Biological Process", fontsize = 20)
ax.set_ylabel("-10log(adjusted p-values)", fontsize = 20)

plt.show()
```
